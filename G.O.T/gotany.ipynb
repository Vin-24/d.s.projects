```python
#here we are changing name attacker_1 to primary_attacker
bt.rename(columns={'attacker_1':'primary_attacker'},inplace= True)
bt.info()
```

```python
#we are changing the name of defender_1 to primary_defender
bt.rename(columns={'defender_1':'primary_defender'},inplace= True)
bt.head()
```

```python
#checking how many time the king attacked
bt['attacker_king'].value_counts()
```

```python
#checking the location in which battle occured
bt['location'].value_counts()
```

```python
#the size of the army of attacker king
sns.set(rc={'figure.figsize':(13,10)})
#sns.set_theme(style="ticks")
sns.set_theme(style="darkgrid")
sns.barplot(x='attacker_king', y='attacker_size', data=bt)
#sns.pointplot(x='attacker_king', y='attacker_size', data=bt, color="blue")
#sns.stripplot(x='attacker_king', y='attacker_size', data=bt, color="blue")
plt.show
```

```python
#army size of the defender
sns.set(rc={'figure.figsize':(13,10)})
sns.stripplot(x='defender_king', y='defender_size', data=bt, color="blue")
plt.show()
```

```python
#differnt types of battle
sns.countplot(x=bt["attacker_king"], hue=bt["battle_type"])
plt.show()
```
