```python
#creating the barplot for countries from where most player come from
plt.figure(figsize=(10,10))
plt.bar(list(fifa['nationality'].value_counts()[0:10].keys()),list(fifa['nationality'].value_counts()[0:10]),color="green")
#first we have taken the country without categorical value using key() so that we can print name then we have taken the nationality with the categorical value this will give the barplot of which country have more number of player
plt.show()
```

```python
#knowing the salary of player
player_salary=fifa[['short_name','wage_eur']] 
#also we have created the sub data of salary of player with there short name
print(player_salary)
```

```python
player_salary.head(10)
```

```python
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
print(player_salary)
```

```python
plt.figure(figsize=(10,10))
plt.bar(list(player_salary['short_name'].value_counts()[0:10].keys()),list(player_salary['wage_eur'].value_counts()[0:10]),color=["darkgreen","forestgreen","green","seagreen","limegreen","mediumseagreen","lime","springgreen","mediumspringgreen","mediumaquamarine"])
plt.show()
#making the barplot for player salary
```
